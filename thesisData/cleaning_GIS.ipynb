```python
from lxml import html
from lxml import etree

i = 0
namelist = []
loclist = []

tree = html.parse("files/2021_loc.html").find("//div[@class='WordSection1']")
XML = html.tostring(tree)

parser = etree.XMLParser(remove_blank_text=True) 
root = etree.XML(XML, parser)

for element in root.iter("*"):
     if element.text is not None and not element.text.strip(): element.text = None

# parser = etree.XMLParser(remove_blank_text=True) 
# root = etree.XML("files/2021_loc.html", parser)

# tree = html.parse("files/2021_loc.html").find("//p[@class='MsoNormal']")
```

```python
import os
from dotenv import load_dotenv

from geopy.geocoders import GoogleV3
from geopy.geocoders import Nominatim
```

```python
for element in root.iter("p"):
     print(str(element.text))
```

```python
def getBorough(long):  
    if 'Queens' in long:
        return 'Queens'
    elif 'Brooklyn' in long:
        return 'Brooklyn'
    elif 'Manhattan' in long:
        return 'Manhattan'
    elif 'Bronx' in long:
        return 'Bronx'
    elif 'Staten Island' in long:
        return 'Staten Island'

def getIntersect(long):    
    long = long.replace('\n', ' ').replace('\r', '').replace('   ', ' ')
    if 'Between' in long:
        long = long.split('Between')[1].strip()
    elif 'of' in long:
        long = long.split('of',1)[1].strip()
    else:
        long = '10 Montieth Street'
        
    return long

def finalize(borough,name,street,intersect):
    if street == 'None':
        location = intersect+','+borough+',NY'
    else:
        location = street+','+borough+',NY'
    print(name)
    print(location)
    namelist.append(name)
    loclist.append(location)
```

```python
i=0
j=0
borough,name,street,intersect = '','','',''

for element in root.iter("p"):
    if i==0: 
        borough = getBorough(element.text)
        i+=1
    elif i==2: 
        name = element.text 
        i+=1
    elif i==3: 
        street = element.text
        i+=1
    elif i==4: 
        if element.text :
            intersect = getIntersect(element.text)
        i+=1
    elif i==5: 
        finalize(borough,name,street,intersect)
        borough,name,street,intersect = '','','',''
        i=0
        j += 1
        print('-------------')
        print(j)
        print('-------------')
    else:
        i +=1
        
        
```
